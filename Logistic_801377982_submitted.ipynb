```python

dfHR_data['PromSpendCategory'] = dfHR_data['PromSpend'].apply(lambda x: 'Over 400' if x > 400 else 'Under 400')

plt.figure(figsize=(10, 6))

# Create a bar plot with custom colors
ax = sns.countplot(data=dfHR_data, x='PromSpendCategory', hue='TargetBuy', palette={0: 'grey', 1: 'blue'})

# Add count annotations on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Count of TargetBuy based on PromSpend Category')
plt.xlabel('PromSpend Category')
plt.ylabel('Count')
plt.show()
```

```python
y = np.array(dfHR_data["TargetBuy"])
X = dfHR_data[["DemAffl", "DemAge", "PromTime", "PromClass_Gold", "PromClass_Platinum", "PromClass_Silver", "DemGender_F", "DemGender_M"]]
```

```python
X.info()
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=44)
```

```python
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
```

```python
model = sm.Logit(y_train, X_train)
result = model.fit()
print(result.summary())
```

```python
X_test["Predicted_Prob"] = result.predict(X_test)
X_test.head()
```

```python
predictions= (X_test["Predicted_Prob"] >= 0.5).astype(int)
predictions
```

```python
from sklearn import metrics

conf_matrix = metrics.confusion_matrix(y_test, predictions)

conf_matrix
```

```python
print("True Positve:",conf_matrix[1][1])
print("True Negative:",conf_matrix[0][0])
print("False Postive:",conf_matrix[0][1])
print("False Negative:",conf_matrix[1][0])
```
